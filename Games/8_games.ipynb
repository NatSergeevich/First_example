**Описание данных**
- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Расшифровка рейтинга ESRB согласно Википедии:

   - `EC` («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры. Некоторые усложнённые развивающие игры могут иметь рейтинг «Everyone»;
   - `E` («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера;
   - `E10+` («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови;
   -  `T` («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики;
   -  `M` («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории;
   -  `AO` («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое, а также сцены с обнажением;
   -  `RP` («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.
   


Данные за 2016 год могут быть неполными.

**Соотнесение выводов по влиянию оценок пользователей и критиков с продажами игр на других платформах**

Для интерпритации данных о влиянии  оценок пользователей и критиков на продажи игр построим диаграммы рассения и рассчитаем корреляцию.

```python
# создадим функцию для построения диаграмм рассеяния и рассчёта корреляции
def corr_analys(platform):
    data = actual_data.query("platform == @platform")
    total_sales = data['total_sales']
    user_score = data['user_score']
    critic_score = data['critic_score']
    
    corr_ts_us = total_sales.corr(user_score)
    corr_ts_cs = total_sales.corr(critic_score)
    
    print(f'Коэффициент корреляции между глобальными продажами и оценкой игроков: {corr_ts_us:.2}')
    print(f'Коэффициент корреляции между глобальными продажами и оценкой критиков: {corr_ts_cs:.2}')
    
    sns.scatterplot(x=total_sales, y=user_score, color='r').set(title='Диаграмма рассеяния - влияние оценок пользователей на уровень продаж',
                                                       xlabel='Глобальные продажи, млн.копий',
                                                       ylabel='Оценка пользователей',                        
                                                    )
    plt.show()
    sns.scatterplot(x=total_sales, y=critic_score).set(title='Диаграмма рассеяния - влияние оценок критиков на уровень продаж',
                                                       xlabel='Глобальные продажи, млн.копий',
                                                       ylabel='Оценка критиков'
                                                    )
    plt.show()
```

```python
# обработаем с помощью циклов платформ из актуального периода
for platform in actual_data['platform'].unique():
    print(f'Анализ для платформы {platform}:')
    corr_analys(platform)
```

Анализ диаграмм рассеяния влияния оценок пользователей и критиков, а также расчёты коэффрициента корреляции среди всех платформ подтвеждают что оценки пользователей влияют на продажи слабо, оценки критиков влияют больше, однако, эта связь не сильна.

## Составление портрета пользователя каждого региона

Для составления портрета пользователя каждого региона (NA, EU, JP) необходимо:
 - исследовать самые популярные платформы (топ-5) и описать различия в долях продаж;
 - исследовать сеамые популярные жанры (топ-5) и пояснить разницу;
 - определить влияет ли рейтинг ESRB на продажи в отдельном регионе?


```python
#выберем нужные нам столбцы с регионами из датасета
markets = data.columns[4:7]
```

Выделим самые популярные платформы  и жанры (топ-5) для каждого региона (NA, EU, JP). Так как необходимо исследовать всего 5 платформ и посмотреть доли, построим круговые диаграммы. Долю не вошедшив топ-5 объединим и тоже покажем на диаграмме

```python
# выведем круговую диаграмму для каждого региона
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('ТОП-5 платформ продаж в регионах')

# выберем из актуальных данных топ 5 самых продаваемых платформ 
for i in range(3):
    df_i = actual_data.groupby('platform')[markets[i]].sum().reset_index()
    df_i = df_i.sort_values(by=markets[i], ascending=False).head(5)
    
    # выделим суммарную долю остальных платформ
    other_sum = actual_data[markets[i]].sum() - df_i[markets[i]].sum()
    df_other = pd.DataFrame({'platform': ['other_platform'], markets[i]: [other_sum]})
    df_i = pd.concat([df_i, df_other])  # обновленный способ конкатенации датафреймов
    
    axs[i].pie(df_i[markets[i]], labels=df_i['platform'], autopct='%1.1f%%')
    
    # выведем аббревиатуру региона   
    axs[i].set_title(markets[i].split('_')[0].upper())
plt.show()
```

## Проверка гипотез.

**5.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**

Сформулируем нулевую и альтернативную гипотезы. Нулевую гипотезу будем формулировать так, чтобы использовать знак равенства. Альтернативную гипотезу сформулируем отталкиваясь от нулевой гипотезы.

- H₀ - средние пользовательские рейтинги платформ Xbox One и PC равны
- H₁ - средние пользовательские рейтинги платформ Xbox One и PC не равны

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(actual_data.loc[actual_data['platform'] =="XOne"]['user_score'], actual_data.loc[actual_data['platform'] =="PC"]['user_score'], equal_var = False,
    nan_policy='omit')
# alpha = зададим значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран
print('p-значение: ', results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
# проверим средние пользовательские рейтинги по платформам
actual_data.query('platform in ["XOne", "PC"]')\
.pivot_table(index='platform', values='user_score', aggfunc=['count', 'mean'])
```

**Вывод:** Не удалось отвергнуть нулевую гипотезу на уровне значимости 5%. Нет достаточных доказательств в поддержку утверждения о том, что средние пользовательские рейтинги платформ Xbox One и PC не равны.

**5.2 Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.**

Сформулируем нулевую и альтернативную гипотезы:

- H₀ - средние пользовательские рейтинги жанров Action и Sports равны
- H₁ - средние пользовательские рейтинги жанров Action и Sports не равны

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(actual_data.loc[actual_data['genre'] =="Action"]['user_score'], actual_data.loc[actual_data['genre'] =="Sports"]['user_score'], equal_var = False,
    nan_policy='omit')
# alpha = зададим значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран
print('p-значение: ', results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
# проверим средние пользовательсие рейтинги по жанрам
actual_data.query('genre in ["Action", "Sports"]')\
.pivot_table(index='genre', values='user_score', aggfunc=['count', 'mean'])
```

**Вывод:** Отвергаем нулевую гипотезу на уровне значимости 5%. Существует достаточно доказательств в поддержку утверждения о том, что средние пользовательские рейтинги жанров Action и Sports не равны.

   сформулированы нулевая и альтернативная гипотезы: H₀ - средние пользовательские рейтинги жанров Action и Sports равны; H₁ - средние пользовательские рейтинги жанров Action и Sports не равны.
     
*2) средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные:*

   сформулированы нулевая и альтернативная гипотезы: H₀ - средние пользовательские рейтинги жанров Action и Sports равны
и H₁ - средние пользовательские рейтинги жанров Action и Sports не равны 

При формулировании нулевых гипотез мы исходим из следующего утверждения - пользовательские рейтинги равны. Были сформулированы двусторонние гипотезы, потому что есть возможное отклонение и в большую, и в меньшую стороны. Для проверки применили t-критерий - это статистический метод, который позволяет сравнивать средние значения двух выборок и на основе результатов теста делать заключение о том, отличаются ли они друг от друга статистически или нет. Выборки являются независимыми. Для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам был применён метод Python: scipy.stats.ttest_ind (array1, array2, equal_var), array1, array2 — массивы, содержащие выборки; equal_var параметр, задающий, считать ли равными дисперсии выборок. Возможно выборки разного размера и/или можно предположить, что дисперсии у генеральных совокупностей разные, был указан "equal_var= False".
